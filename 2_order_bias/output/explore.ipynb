```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your CSV
df = pd.read_csv("new_0.2_order_bias_mistral-small_cot_en/new_0.2_order_bias_mistral-small_cot_en.csv")
# df = df.head(3000)

# Filter correct predictions (Predicted == RotationPosition)
df_correct = df[df["RotationPosition"] == df["Predicted"]]

# Count correct answers at each rotation position
correct_counts = df_correct["RotationPosition"].value_counts().sort_index()

# Plot the results
plt.figure(figsize=(8, 5))
plt.bar(correct_counts.index, correct_counts.values)
plt.xlabel("Rotation Position")
plt.ylabel("Number Correct")
plt.title("Correct Predictions by Rotation Position")
plt.grid(True, axis='y')
plt.xticks(correct_counts.index)
plt.tight_layout()
plt.show()

std_dev = correct_counts.std()
range_val = correct_counts.max() - correct_counts.min()
print(correct_counts)
print(f"Standard Deviation of Answer Placement: {std_dev} and Range of Rotation Position: {range_val}")
print(f' % correct: {df_correct.shape[0] / df.shape[0]} and number correct: {df_correct.shape[0]} / {df.shape[0]}')
```
