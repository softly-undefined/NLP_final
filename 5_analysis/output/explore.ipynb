# Explore

This code contains the beginning of the exploration that led to the generation of the different figures used in the paper.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('fullmetrics.csv')
df.head(12)
```

# Begin the Analysis

## Overall Accuracy

```python
df['color'] = df['model_name'].apply(lambda x: '#FFC5B7' if 'zh' in x else '#A1B9E2')
df_sorted = df.sort_values(by='overall_accuracy', ascending=False)

plt.bar(df_sorted['model_name'], df_sorted['overall_accuracy'], color=df_sorted['color'])
plt.xlabel('Model Name')
plt.xticks(rotation=90)
plt.ylabel('overall_accuracy')
plt.title("overall_accuracy values")

plt.show()
```

```python
df_sorted = df.sort_values(by='overall_accuracy', ascending=False)
df_sorted = df_sorted[df_sorted['model_name'].str.contains('en')]
plt.bar(df_sorted['model_name'], df_sorted['overall_accuracy'])
plt.xlabel('Model Name')
plt.xticks(rotation=90)
plt.ylabel('Overall Accuracy')

plt.show()
```

```python
# Chinese overall_accuracy
df_sorted = df.sort_values(by='overall_accuracy', ascending=False)
df_sorted = df_sorted[df_sorted['model_name'].str.contains('zh')]
plt.bar(df_sorted['model_name'], df_sorted['overall_accuracy'])
plt.xlabel('Model Name')
plt.xticks(rotation=90)
plt.ylabel('Overall Accuracy')

plt.show()
```
